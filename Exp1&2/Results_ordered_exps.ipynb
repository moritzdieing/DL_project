```python


# Simulate data for 5 experiments, each with 5 executions
num_experiments = 5
num_executions = 5
num_epochs = 10

# Function to plot mean and standard deviation for an experiment
def plot_experiment(ax, experiment_name, train_acc, test_acc):
    epochs = np.arange(1, num_epochs + 1)

    # Calculate mean and standard deviation for training and test accuracy
    train_mean = train_acc.mean(axis=0)
    train_std = train_acc.std(axis=0)
    test_mean = test_acc.mean(axis=0)
    test_std = test_acc.std(axis=0)

    # Plot training accuracy with shaded region for standard deviation
    ax.plot(epochs, train_mean, label="Train Accuracy", color="blue")
    ax.fill_between(epochs, train_mean - train_std, train_mean + train_std, color="blue", alpha=0.2)

    # Plot test accuracy with shaded region for standard deviation
    ax.plot(epochs, test_mean, label="Test Accuracy", color="orange")
    ax.fill_between(epochs, test_mean - test_std, test_mean + test_std, color="orange", alpha=0.2)

    # Set plot labels and title
    ax.set_title(experiment_name, fontsize=14)
    ax.set_xlabel("Epoch", fontsize=14)
    ax.set_ylabel("Accuracy", fontsize=14)
    ax.legend(fontsize=12)
    ax.grid(True)
```

```python
"""
Plot train and test acc. for samples when sorted in DESCENDING order. This
is for 5 runs and only experiment 0. (normal permutation) --> similar behaviour in other runs.
"""

# Create subplots for each experiment
fig, axes = plt.subplots(1, num_experiments, figsize=(20, 5), sharey=True)

# Plot each experiment
for i, (experiment_name, results) in enumerate(data_desc.items()):
    plot_experiment(axes[i], experiment_name, results["train_acc"], results["test_acc"])

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import statistics

def run_average(perm, path, exp_names, desc, verbose):
  total = {}
  for idx_1, exp_name in enumerate(exp_names):
    averages = [0, 0, 0, 0, 0]
    if verbose: print(f"experiment {exp_name}")

    for run in range(5):
      if verbose: print(f"Run: {run}")
      inner_averages = []
      learn_speed = torch.load(f'{path}/ls_{exp_name}_{desc}_keep_run{run}.pth')
      count = [torch.bincount(learn_speed[i].int()) for i in range(5)]

      # average learning speed pro task in run:
      for idx_2, t in enumerate(count):
        avg = 0
        for idx_in, el in enumerate(t):
          avg += el * idx_in
        if verbose: print(f"Learned Task {idx_2} (actual task {perm[idx_1][idx_2]}) avg. learning speed: {avg/t.sum()}")
        inner_averages.append(avg/t.sum())

      # compute speed average for this run
      averages[run] = float(sum(inner_averages) / len(inner_averages))
    for pr_idx, avg in enumerate(averages):
      if verbose: print(f"Run {pr_idx} average: {avg}")

    print(averages)
    total[exp_name] = (statistics.mean(averages), statistics.stdev(averages))
  return total
```
